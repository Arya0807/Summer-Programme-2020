```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
trainset = pd.read_csv('Trainfile.csv')
testset = pd.read_csv('Testfile.csv')
```

```python
trainset.describe()
```

```python

X_train=trainset.drop(['Utilities','Condition2', 'Exterior1st','FireplaceQu','ExterCond','MiscFeature', 'SaleType','SalePrice'],axis=1)
X_test=testset.drop(['Utilities','Condition2', 'Exterior1st','FireplaceQu','ExterCond','MiscFeature', 'SaleType','SalePrice'],axis=1)
'''The above columns were dropped because they had categories in test data which were not present in train data so encoding was not possible'''

Y_train = trainset.iloc[:,[80]]
Y_test = testset.iloc[:,[80]]

s = (X_train.dtypes == 'object')
cat_cols = list(s[s].index)

print("Categorical variables:")
print(cat_cols)
```

```python
X_train_cats=X_train[cat_cols]
X_test_cats=X_test[cat_cols]
X_train_nums=X_train.drop(['MSZoning', 'Street', 'Alley', 'LotShape', 'LandContour', 'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl',  'Exterior2nd', 'MasVnrType', 'ExterQual', 'Foundation', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual', 'Functional',  'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'PavedDrive', 'PoolQC', 'Fence',  'SaleCondition'],axis=1)
X_test_nums=X_test.drop(['MSZoning', 'Street', 'Alley', 'LotShape', 'LandContour', 'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl',  'Exterior2nd', 'MasVnrType', 'ExterQual',  'Foundation', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual', 'Functional',  'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'PavedDrive', 'PoolQC', 'Fence', 'SaleCondition'],axis=1)
```

```python
#Taking care of missing data
from sklearn.preprocessing import Imputer
imputer = Imputer(missing_values = 'NaN', strategy = 'mean', axis = 0)
imputer = imputer.fit(X_train_nums)
X_train_nums = imputer.transform(X_train_nums)
X_test_nums = imputer.transform(X_test_nums)
X_train_nums=pd.DataFrame(X_train_nums)
X_train_nums.columns=X_train.drop(['MSZoning', 'Street', 'Alley', 'LotShape', 'LandContour', 'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl',  'Exterior2nd', 'MasVnrType', 'ExterQual', 'Foundation', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual', 'Functional',  'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'PavedDrive', 'PoolQC', 'Fence',  'SaleCondition'],axis=1).columns
X_test_nums=pd.DataFrame(X_test_nums)
X_test_nums.columns=X_test.drop(['MSZoning', 'Street', 'Alley', 'LotShape', 'LandContour', 'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl',  'Exterior2nd', 'MasVnrType', 'ExterQual', 'Foundation', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual', 'Functional',  'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'PavedDrive', 'PoolQC', 'Fence',  'SaleCondition'],axis=1).columns

from sklearn.impute import SimpleImputer
imputer1 = SimpleImputer(strategy='most_frequent')
X_train_cats= pd.DataFrame(imputer1.fit_transform(X_train_cats))
X_test_cats = pd.DataFrame(imputer1.transform(X_test_cats))
X_train_cats.columns=X_train[cat_cols].columns
X_test_cats.columns=X_test[cat_cols].columns
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train_nums)
X_train_nums = scaler.transform(X_train_nums)
X_test_nums = scaler.transform(X_test_nums)
scaler.fit(Y_train)
Y_train = scaler.transform(Y_train)
Y_test = scaler.transform(Y_test)
Y_train=pd.DataFrame(Y_train)
Y_train.columns=trainset.iloc[:,[80]].columns
X_train_nums=pd.DataFrame(X_train_nums)
X_train_nums.columns=X_train.drop(['MSZoning', 'Street', 'Alley', 'LotShape', 'LandContour', 'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl',  'Exterior2nd', 'MasVnrType', 'ExterQual', 'Foundation', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual', 'Functional',  'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'PavedDrive', 'PoolQC', 'Fence',  'SaleCondition'],axis=1).columns
X_test_nums=pd.DataFrame(X_test_nums)
X_test_nums.columns=X_test.drop(['MSZoning', 'Street', 'Alley', 'LotShape', 'LandContour', 'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl',  'Exterior2nd', 'MasVnrType', 'ExterQual', 'Foundation', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual', 'Functional',  'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'PavedDrive', 'PoolQC', 'Fence',  'SaleCondition'],axis=1).columns
```

```python
#Label and One Hot encoding
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
for col in cat_cols:
    X_train_cats[col] = labelencoder.fit_transform(X_train_cats[col])
    X_test_cats[col] = labelencoder.transform(X_test_cats[col])
print(pd.DataFrame(X_test_cats))    
from sklearn.preprocessing import OneHotEncoder
OH_encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
OH_cols_train = pd.DataFrame(OH_encoder.fit_transform(X_train_cats[cat_cols]))
OH_cols_test = pd.DataFrame(OH_encoder.transform(X_test_cats[cat_cols]))    
```

```python
#Correlation matrix for numerical columns
X_train_nums=pd.concat([X_train_nums,Y_train],axis=1)
cor = X_train_nums.corr()
cor.head()
```

```python
#Correlation matrix for categorical columns
OH_cols_train=pd.concat([OH_cols_train,Y_train],axis=1)
cor1 = OH_cols_train.corr()
cor1.head()
```

```python
X_Train_nums=X_train_nums[['LotFrontage', 'LotArea', 'OverallQual', 'YearBuilt', 'YearRemodAdd',
       'MasVnrArea', 'BsmtFinSF1', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF',
       'GrLivArea', 'FullBath', 'HalfBath', 'TotRmsAbvGrd', 'Fireplaces',
       'GarageYrBlt', 'GarageCars', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF','OverallCond', 'KitchenAbvGr', 'EnclosedPorch']]
X_Test_nums=X_test_nums[['LotFrontage', 'LotArea', 'OverallQual', 'YearBuilt', 'YearRemodAdd',
       'MasVnrArea', 'BsmtFinSF1', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF',
       'GrLivArea', 'FullBath', 'HalfBath', 'TotRmsAbvGrd', 'Fireplaces',
       'GarageYrBlt', 'GarageCars', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF','OverallCond', 'KitchenAbvGr', 'EnclosedPorch']]
X_Train_cats=OH_cols_train[[40, 41,99,106, 108, 112, 116, 146, 158, 160, 175, 203,104, 109, 111, 119, 127, 150, 161, 174, 177]]
X_Test_cats=OH_cols_test[[40, 41,99,106, 108, 112, 116, 146, 158, 160, 175, 203,104, 109, 111, 119, 127, 150, 161, 174, 177]]
X_Train=pd.concat([X_Train_nums,X_Train_cats],axis=1)

X_Test=pd.concat([X_Test_nums,X_Test_cats],axis=1)
#
X_Train = np.append(arr = np.ones((1138,1)).astype(int),values = X_Train, axis = 1 )
X_Test = np.append(arr = np.ones((322,1)).astype(int),values = X_Test, axis = 1 )
```

```python
#Linear regression model
X_Train_trans = X_Train.transpose()
Theta = np.dot(np.dot(np. linalg. inv(np.dot(X_Train_trans,X_Train)),X_Train_trans),Y_train)
Y_pred = np.dot(Theta.transpose(),X_Test.transpose())
Y_pred=scaler.inverse_transform(Y_pred)
Y_test=testset.iloc[:,[80]]
Y_test=np.asarray(Y_test)
Y_pred.reshape(-1,1)
```

```python
#MAPE
i=0
error=0
while (i<len(Y_test)):
    error=error+(np.abs((Y_pred[0][i]-Y_test[i])/Y_test[i]))
    i=i+1
print((error*100/len(Y_test)))    
```

```python
print("Accuracy=")
print(100-int(error*100/len(Y_test)))
```
